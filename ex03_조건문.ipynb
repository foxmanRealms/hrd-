# 조건문

### if 문
- 조건식이 참일 경우에 실행문장을 실행한다



```python
if True:
    print("실행문장 실행")
```

```python
if False:
    print("실행문장 실행")
```

```python
#들여쓰기의 중요성!!!!!
if False:
    print("실행문장1")
print("실행문장2")
```

### if else 문

- 조건식이 참일때는 if 실행문장 실행
- 그 외의 경우에는 else 실행문장이 실행
- ** else는 if와 같은 선상에 써주기! **
- else는 따로 조건식을 적어주지 않아도 된다(참이 아닐 경우 실행문장을 실행)

```python
if False:
    print('실행문장1')
else:
    print('실행문장2')
    
```

조건문 if else 예제


```python
money =1000

if money>=10000:
    print('택시를 탄다~!!')
else:
    print('버스를 탄다')
```

조건문 if else 예제


```python
num=int( input('정수 입력 >> ') )
if num%3==0 and num%5==0:
    print('3과 5의 배수입니다')
else:
    print('3과 5의 배수가 아닙니다')
```

조건문 elif 예제


```python
num1=int( input('첫번째 정수 입력 >> '))
num2=int( input('두번째 정수 입력 >> '))
if num1>num2 :
    print('첫번째 정수가 더 큽니다')
elif num1<num2:
    print('두번째 정수가 더 큽니다')
elif num1==num2:
    print('두 수가 똑같습니다')
```

```python
#1. num1,num2 입력받기
num1=int( input('첫번째 정수 입력 >> '))
num2=int( input('두번째 정수 입력 >> '))
#2. 조건문
if num1>num2 :
    print('첫번째 정수가 더 큽니다')
elif num1<num2:
    print('두번째 정수가 더 큽니다')
else:
    print('두 수가 똑같습니다')
```

조건문 elif 예제  Score


```python
score =int( input('점수 입력 >> ' ))
if score>=90 and score<=100:
    print(f'{score} 점은 A학점 입니다')
elif score>=80 and score<90:
    print(f'{score} 점은 B학점 입니다')
elif score>=70 and score<60:
    print (f'{score} 점은 C학점 입니다')
elif score>=60 and score<70:
    print(f'{score} 점은 D학점 입니다')
elif score<60:
    (f'{score} 점은 F학점 입니다')
```

```python
#1. score 변수에 input 함수 사용하여 점수 입력받기
score =int( input('점수 입력 >> ' ))
#2. 학점판별 조건문(조건을 여러개 판별할 수 있는 elif문)
if score>=90 and score<=100:
    result='A'
elif score>=80 and score<90:
    result='B'
elif score>=70 and score<60:
    result='C'
elif score>=60 and score<70:
    result='D'
elif score<60:
    result='F'
#3. 결과 출력
print(f'{score} 점은 {result}학점 입니다')
```
