# while문
- 반복 횟수가 명확하지 않을 때 사용한다!
- 조건식이 참일 경우에 실행된다!
- while 조건식 : 
- (들여쓰기) 실행문장
- 반복문을 멈추고 싶을 때는 break키워드 사용

```python
# 반복문 실습
num=0
while num<10:
    print('파이썬 최고 짱짱~~!!')
    num+=1
```

while문 예제


```python


while True:
    num1=input('첫 번째 정수 입력 >> ')
    num2=input('두 번째 정수 입력 >> ')
    print('두 정수의 합 : '+str(num1+num2) )
    if num1==0 and num2==0:
        break
print('프로그램이 종료되었습니다.')
    
        
    
```

```python
#2개의 정수를 입력받기
while True:
num1=int( input('첫 번째 정수 입력 >> ') )
num2=int( input('두 번째 정수 입력 >> ') )
result=num1+num2

if num1==0 and num2==0:
    print('프로그램이 종료되었습니다.')
    break
else:
    print(f'두 정수의 합 : {result}')
    
    고장나있음 확인하기
```
