### 문자열 인덱싱과 슬라이싱
- 인덱싱 : 문자열 안에서 내가 가져오고 싶은 데이터 '하나'만 가져올 때 사용한다
- 슬라이싱 : 문자열 안에서 범위를 주고 잘라올 때 사용한다

```python
name = 'My name is JM'
```

```python
#인덱싱 방법 > 변수명[인덱스번호]
#대문자 M 가져오기
print(name[0])
#대문자 J가져오기(음수인덱스 사용)
print(name[-2])
```

```python
name = 'My name is JM'
```

```python
#슬라이싱 방법 > 변수명[시작인덱스:끝인덱스+1]  ** 내가 가져오고 싶은 인덱스+1
#is 슬라이싱 해오기
print(name[8:10])
#'name' 슬라이싱 해오기(음수인덱스 활용)
print(name[-10:-6])
```

```python
#처음부터 가져오고 싶을 때는 콜론 앞 인덱스 써주지 않아도 된다!
#'My name'가져오기
print(name[:7])
```

```python
#끝까지 가져오고 싶을 때는 콜론 뒤 인덱스 써주지 않아도 된다!
#'is JM'가져오기
print(name[-5:])
```

```python
#모든 문자 가져오기
print(name[:])
```

문자열 슬라이싱 예제


```python
yd_info="2002 0412 Hot"

print('연도: '+yd_info[:5])
print('월일: '+yd_info[-8:-4])
print('날씨: '+yd_info[-3:])
```

선생님 풀이

```python
yd_info="2002 0412 Hot"
year =yd_info[:4]
day =yd_info[5:9]
weather=yd_info[10:]
```

```python
#파이썬에서 문자열 합칠때는 +기호 사용
print('연도: '+year)
print('월일: '+day)
print('날씨: '+weather)
```

### 포매팅
- 문자열 안에서 특정한 값을 삽입해야 할 때 사용한다
    - 3가지의 포매팅방법
        - %기호 포매팅
        - format 함수 포매팅
        - f문자열 포매팅

```python
#월과 일을 바꿔주고 싶다면??
s5='오늘은 4월 12일 입니다!'
s5
```

```python
month = 4
day = 12
s6='오늘은'+month +'월'+day+'입니다'
```

```python
# %기호 포매팅 실습 -> %기호를 활용해서 정수를 대입한다
month = 4
day = 15
s7='오늘은 %d월 %d일 입니다'%(month,day)
print(s7)
```

문자열 포매팅


```python
#format 함수를 활용한 포매팅 실습

month = 4
day = 12
s8='오늘은 {}월 {}일 입니다'.format(month,day)
print(s8)
```

```python
# f문자열 포매팅 방법을 사용한 포매팅 실습
month = 4
day = 12
s9=f'오늘은 {month}월 {day}일 입니다'
print(s9)
#직관적이라는 장점이 있다.
```

변수실습 예제


```python
x=100
y=200
add = x+y
res=f'{x}와 {y}의 합은 {add}입니다'
print(res)
```

문자열함수


```python
#count > 특정 데이터의 개수를 알려줌
s10='my name is younghwa'
print(s10.count('a') )

#index() > 특정문자의 위치를 알려준다
print(s10.index('i'))

#replace(바꿔주고 싶은 문자, 바뀔 문자)
print(s10.replace('younghwa','jisu'))
```
