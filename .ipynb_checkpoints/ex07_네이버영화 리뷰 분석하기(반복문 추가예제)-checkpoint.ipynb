##### 1. 전체 리뷰 개수 확인하기

```python
#리스트 안의 리스트 개수가 몇개인지 세주기
len(data)
```

##### 2. 리뷰길이의 평균값 구하기

```python
total_len=0

for r in data:   
    total_len += len(r[0])
    avg=total_len/len(data)
#print(avg)
print(f'리뷰의 평균 길이는"{avg}"입니다.')        
    
```

```python
total=0
for r in data:
    total+=len(r[0] )  #누적해주기
print(f'리뷰의 평균 길이는 "{total/len(data)}"입니다')
```

#### 3. 긍정리뷰/부정리뷰 개수 구하기

```python
data
pos,nega=0,0

for r in data:
    if r[1]==0:
        nega+=1
    elif r[1]==1:
        pos+=1
#print(pos)
#print(nega)
print(f'긍정 "{pos}"개, 부정 "{nega}"개')
```

```python
r_list=[]
for r in data:
    r_list.append(r[1])
n= r_list.count(0)
p=r_list.count(1)
print(f"긍정 {p}개")
print(f"부정 {n}개")
```

#### 4. 리뷰 최대길이 값 구하기 / 그 리뷰 내용 출력


```python
data
max_r,idx=0,0
for r in data:
    if max_r<len(r[0]):
        max_r=len(r[0])
        idx=r[0]
#print(max_r)
print(f'리뷰 최대길이는 "{max_r}"입니다.')
print(idx)
```

```python
max_len=0
text=''
for r in data:
    r_len=len(r[0])
    if max_len<r_len:
        max_len=r_len
        text=r[0]
print(max_len)
print(text)
```

검색데이터가 포함된 리뷰 개수



```python
cnt=0
s= input('검색할 데이터 입력 >> ')

for r in data:
    if s in r[0]:
        cnt+=1
print(f'"{s}"이 포함된 리뷰는 {cnt}개 입니다')
```
