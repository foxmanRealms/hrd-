### 리스트
- 많은 양의 자료를 하나로 묶어서 관리할 때 사용
- 순서가 있는 자료 형태이다
- 자료형의 구분 없이 하나로 묶을 수 있다
- [] 대괄호를 사용하여 생성할 수 있다
- 추가, 수정, 삭제가 가능하다 (튜플 과의 차이점)

```python
#리스트 생성하기
list1=[0,1,2,3,4,5]
list1
#출력 결과가 대괄호 안에 들어있다면 리스트 구조이다!
```

```python
type(list1)
```

```python
#서로 다른 자료형태도 묶어줄수 있다 
list2=['영화',313]
list2
```

```python
# 리스트 안에 또다른 리스트를 넣어줄 수 있다
list3=['영화', 313,[0,1,2,3] ]
list3
```

### 리스트 슬라이싱

리스트 값 추가 


```python
f_list=['사과','포도',['수박','멜론'],'복숭아','딸기','오렌지']
a,b,c =f_list[0], f_list[2][0], f_list[-1]
choice_list=[]

choice_list.append(a)
choice_list.append(b)
choice_list.append(c)

print(choice_list)
```

```python
f_list=['사과','포도',['수박','멜론'],'복숭아','딸기','오렌지']
#과일꺼내와서 변수에 담아주기
apple =f_list[0]
watermelon= f_list[2][0]
orange= f_list[-1]
#초이스리스트 생성해주기
choice_list=[]
#append()
choice_list.append(apple)
choice_list.append(watermelon)
choice_list.append(orange)
#출력
print(choice_list)
```

##### 2. 원하는 인덱스 위치에 값 추가하기(insert() )

```python
list5=[0,1,2,3,4,5]
list5
```

```python
#리스트명.insert(인덱스번호, 값) --> 자동으로 인덱스는 밀린다
list5.insert(3,6)
print(list5)
```

insert 실습


```python
music_list = [['가나다라','박재범'], ['LOVE DIVE', '아이브'], ['TOMBOY','아이들']]
loc=int( input('인덱스 위치 입력 >> '))
song=input('노래 제목 입력 >> ')
singer=input('가수 입력 >> ')

addList=[]
addList.append(song)
addList.append(singer)

addList
                                                      
music_list.insert(loc,addList)
```

```python
#1. loc(숫자),song, singer 입력받기
music_list = [['가나다라','박재범'], ['LOVE DIVE', '아이브'], ['TOMBOY','아이들']]
loc=int( input('인덱스 위치 입력 >> '))
song=input('노래 제목 입력 >> ')
singer=input('가수 입력 >> ')

#2.뮤직리스트에 insert해주기 (2번째자리)
music_list.insert(loc,[song,singer])
#3.뮤직리스트 출력
print(music_list)
```

##### 3. 리스트에 있는 값 수정하기

```python
list6=[0,1,2,3,4,5]
list6
```

```python
#리스트명[넣어주고싶은 인덱스]=값     인덱스를 뽑아와서 대입해주는 식의 느낌
#3을 6으로 수정하기

list6[3]=6
list6
```

```python
#리스트 안의 리스트에 값 추가하기 
list5[3].insert(2,4)
```

```python
# 1과 2를 7로 수정하기
list6[1:3]=7
# 오류가 난다
# 왜? 슬라이싱된 결과가 리스트 형태이기 때문에!! 
list6
```

```python
#각각 값을 추가하고 싶을 때는 리스트 안에 그 개수만큼 넣어주면 된다.
#1과 2가 하나의 7로 수정됨
list7=[0,1,2,3,4]
list7[1:3]=[7,7]
list7
```

##### 4. 리스트 삭제(del,remove)

```python
list8=['영화','고은','설미','다민','시윤','진관']
list8
```

```python
#del 리스트이름[삭제하고 싶은 인덱스번호]
del list8[0]
list8
```

```python
del list8[1:3]
list8
```

```python
#리스트이름remove(값) -- 인덱스값이 아니라 실제 데이터값을 적어준다
list8.remove('시윤')
list8
```

```python
#동일한 값이 여러개가 있다면 제일 앞 데이터만 삭제된다!
list9=[0,0,0,0,0,1,2,3,4]
list9.remove(0)
list9
```

##### 5. 정렬(sort,reverse)

```python
#정렬: 리스트이름.sort()
list10=[2,15,9,20,62,69,30]
list10.sort() #오름차순이 기본설정이다
list10
```

```python
#reverse() : 현재 리스트의 값들을 뒤집어서 보여준다.
list11=[2,15,9,20,62,69,30]
list11.reverse()
list11
```

```python
#내림차순 하는 방법 : sort() 해준 다음에 reverse() 해주기
list11.sort()
list11.reverse()
list11
```

##### 6. 기타함수들
- index(값) : 원하는 값의 위치를 알려준다
- pop() : 리스트의 마지막값을 출력해주고 리스트에서 제거한다
- len(리스트이름) : 리스트의 길이를 알려준다 ( length의 약어) -> 요소의 개수를 알려준다

```python
#리스트이름.index(값)
list12=['지수','솔아','민근','정명']
list12.index('솔아')
```

```python
#len(리스트이름)
len(list12)
```

```python
#리스트이름.pop() -> 제일 마지막 값을 보여준다
list12.pop()
list12
```

# in, not in
- in : 찾고자 하는 값(x)가 포함되어 있으면 True를 출력
- not in : 찾고자 하는 값(x)가 포함되어 있지 않으면 True 출력 


in  // not in 연산자 예제 
별이 35개 있다


```python
lst=['딸기','바나나','수박','체리','포도']
scan= input('검색할 문자를 입력하세요 >> ')
if scan in lst:  
    idx= str(lst.index(scan) )
    print( scan+'은(는) 리스트의 '+idx+'번째 인덱스에 들어있습니다')
else:
    print(scan+'은(는) 리스트에 들어있지 않습니다')
```

```python

name='Hi,My name is Younghwa'
scan=input('검색할 문자를 입력하세요 >> ')
print('*'*35)
if scan in name:
    cnt=str(name.count(scan) )
    print(scan+'은(는) 문자열에 '+cnt+ '번 들어있습니다')
else:
    print(scan+'은(는) 문자열에 들어있지 않습니다')

print('*'*35)
```

```python
name='Hi,My name is Younghwa'
#1. 검색할 문자 입력받기
input_s=input('검색할 문자를 입력하세요 >> ')
#2. 조건문으로 판별해주기 / 개수 세서 출력해주기 (count() )
# 변수명.count('검색하고 싶은 문자')
if input_s in name:
    
    print( f'{input_s}은(는) 문자열에 {name.count(input_s)}번 들어있습니다')
else:
    print( f'{input_s}은(는) 문자열에 들어있지 않습니다!')
```
