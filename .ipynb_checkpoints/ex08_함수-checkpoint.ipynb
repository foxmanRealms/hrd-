```python
# 변수 두개를 받아서 더한 값을 출력해주는 함수(num_sum) 정의 하기
def num_sum(num1,num2):
    result=num1+num2
    return result
#아직까지는 함수 정의만 해준 것이다!!    
```

```python
# 함수 사용하기! (호출)
num_sum(2,10)
```

함수예제 2 


```python
num1=int( input('첫번째 정수 입력 >> '))
num2=int( input('두번째 정수 입력 >> '))
op=input('연산자 입력(+,-) >> ')

def cal(n1,n2,op):
    result=0
    if op=='+':     
        cal=num1+num2
        return cal
    elif op=='-':
        cal=num1-num2
        return cal
    else:
        print('잘못 입력하셨습니다.')

result=cal(num1,num2,op)
print(f'결과 : {result}')
```

```python
# 두 수를 입력받고 원하는 연산을 수행하는 함수 정의하기
num1=int( input('첫번째 정수입력 >> '))
num2=int( input('두번째 정수입력 >> '))
op=input('연산자 입력 (+,-) ')
result=cal(num1,num2,op)
print(result)
```

### 가변매개변수
- 함수를 호출할 때 매개변수의 개수가 정확히 몇개인 지 모를 때 사용한다
- 사용자에게 매개변수의 개수를 정할 수 있도록 해준다
- 매개변수의 결과물은 tuple에 담긴다!
- 매개변수 앞에 * 해준다


```python
def print1(*args):
    print(args)
```

```python
print1(1,2,3,4,5,6,8)
```

가변매개변수 예제 


```python
def add(*args):
    print( sum(args) )
    
```

```python
add(6,5,4,3,2,1)
```

```python
def add(*args):
    result=0
    for i in args:
        result+=i
    return result
    
```

```python
add(1,2,3,4,5,6,7,8)
```

### 함수의 결과 값은 언제나 하나이다!!

```python
#함수의 리턴값이 여러개일때
def add_sub(num1,num2):
    return num1+num2,num1-num2
```

```python
#튜플의 형태로 출력이 된다
add_sub(10,7)
```

```python
#따로 꺼내서 보여주고 싶다면??
add, sub=add_sub(10,7)
print(add)
print(sub)
```

#### 함수 기본값 설정

```python
# **(지수연산자)를 사용하는 함수 생성
def power_num(num1,power=2):  #기본값 설정
    return num1**power
```

```python
power_num(2,3)
```

```python
#설정된 기본값으로 계산이 된다!
power_num(3)
```

### 딕셔너리 형태의 가변 매개변수


```python
#딕셔너리 자료구조를 print해주는 함수를 생성해보자
def print_str(**kwargs): #keywored arguments의 약자!
    print(kwargs)
```

```python
#가변매개변수(**) 호출할 때 짝을 지어준 형태로 넣어줘야 한다
print_str(one=1,two=2,three=3)
```
