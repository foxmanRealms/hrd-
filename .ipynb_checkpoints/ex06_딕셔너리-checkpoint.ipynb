# 딕셔너리(dictionary)
- key : value 한쌍으로 이루어진 자료구조
- 순서가 없는 형태이다(인덱스가 존재하지 않는다)
- { }중괄호를 사용하여 생성한다
- key (불변값) : value (변하는 값)

```python
#딕셔너리 생성하기
dic1 = {'이름':'최영화', '나이':'20', '전화번호':'01012345678'}
dic1
```

```python
#타입확인하기
type(dic1)
```

```python
#key값이 중복된다면?
dic2={'a':1,'b':2,'a':3}
dic2
#가장 마지막에 넣으준 값만 인정된다!!
```

```python
#자료 가져오기(key값 사용하여)
dic1['이름']
```

```python
#데이터 수정하기
dic1['이름']='송다민'
dic1
```

```python
#딕셔너리에 새로운 데이터 추가하기
dic1['생일']='7월 22일'
dic1
```

딕셔너리 실습


```python
dic_test={'노래제목':'아무노래'}


dic_test['가수']='지코'
dic_test['날짜']='2020.01.13'
dic_test
```

-딕셔너리 안의 값 삭제하기

```python
del dic_test['날짜']
dic_test
```

- 딕셔너리 데이터 가져오기

```python
#key값을 불러서 가져오기!
dic_test['노래제목']
```

```python
#get() 기능으로 가져오기!
dic_test.get('노래제목')
```

[key값] vs .get(key값) 차이점!

```python
#[key값] 가져올때, 없는 키값을 넣으면 오류가 발생
dic_test['날짜']
```

```python
# .get(key값)을 가져오면 없는 값이라고 알려준다!
print(dic_test.get('날짜') )
```

```python
#딕셔너리 안의 딕셔너리
data1={'김민근':{'나이':'20+10','성별':'남자'},
       '윤솔아':{'나이':'20*12','성별':'여자'} }
data1
```

```python
#딕셔너리 안에 있는 key값만 가져오기
data1.keys()
```

```python
type(data1.keys() )
```

```python
#딕셔너리 안에 있는 value값들만 가져오기
data1.values()
```

```python
#for문 사용해서 딕셔너리 사용하기
for key in data1.keys():
    print (key)
```

```python
for value in data1.values():
    print (value)
```

```python
#key값과 value값을 둘다 가져오는 방법
data1.items()
```

```python
for key,value in data1.items():
    print(key,value)
```

```python
#자료구조 안에 해당하는 데이터가 존재하는지 알려주는 기능
'윤솔아' in data1
```

```python
'신지수' in data1
```

```python
'윤솔아' not in data1
```

### 딕셔너리 최종 실습

```python
score_dic={'최고은':{'파이썬':100,'스프링':90,'자바':80},
          '정정용':{'파이썬':50,'스프링':100,'자바':100},
          '윤진관':{'파이썬':90,'스프링':50,'자바':80}
          }
```

```python
score_dic
```

```python
for i in score_dic[i]["파이썬"]:
    print
```

```python
score_dic['최고은']['파이썬']+score_dic['정정용']['파이썬']+score_dic['윤진관']['파이썬']
```

```python
score_dic={'최고은':{'파이썬':100,'스프링':90,'자바':80},
          '정정용':{'파이썬':50,'스프링':100,'자바':100},
          '윤진관':{'파이썬':90,'스프링':50,'자바':80}
          }
```

```python
#keys()
score_dic.keys()
```

```python
score_dic['최고은']['파이썬']
```

```python
p,j,s =0,0,0

for name in score_dic.keys():
    p +=score_dic[name]['파이썬']
    j +=score_dic[name]['자바']    
    s +=score_dic[name]['스프링']    
```

```python
result_dic={'파이썬 점수':p,'자바 점수':j, '스프링 점수':s}
result_dic
```
