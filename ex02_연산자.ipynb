산술연산자 예제


```python
time = int( input("시간 입력 >> ") )
hour= time//60//60
minute = time%3600//60
second = time%60

print( "{}시간 {}분 {}초".format(hour,minute,second) )
```

```python
#1. time 변수에 입력받기
time = int( input("시간입력 >> ") )
#2. 시간계산 해주기
hour=time//3600   #1분 60초 / 1시간 60분 
minute = time%3600//60 
second = time%60

#3. 출력
print(f"{hour}시간 {minute}분 {second}초" )
```

### 논리연산자
- not : 논리를 부정해주는 연산자(not 변수)
- and : 둘 다 참일 때 참을 실행해준다
- or : 둘 중 하나만 참이어도 참을 실행해준다

```python
a =10
b=7
print(not a==b)
print(a==b and a>b)
print(a==b or a>b)
```

3항 연산자 실습 예제


### 삼항연산자
- a if 조건식 else b
- 조건식의 결과가 참 이었을 때 a 실행
-   결과가 거짓 이었을때 b 실행

```python
num= input("정수 입력 >> ") 
print(num +" 는(은) 홀수입니다" if int(num)%2==1 else num +" 는(은) 짝수입니다" )
```

```python
age=19
#조건식 : age 20미만이라면 '미성년자 입니다' 20이상이라면 '성인입니다'

'성인입니다' if age >=20 else '미성년자 입니다'
```

```python
#1. 숫자입력받기 
num = int(input("정수입력 >> ") )
#2 삼항연산자를 활용하여 참 거짓 판별하기
result = "짝수" if num%2==0 else "홀수" 
#3. 출력
print(f"{num}는(은) {result}입니다.")
```
